```python
#we need not to include none
df_new=df[df.Milk!="None"]
```

```python
df_new
```

```python
from statsmodels.formula.api import ols
```

```python
from statsmodels.stats.anova import anova_lm
```

```python
formula="Calories~C(Milk)"
```

```python
model=ols(formula,df_new).fit()
anova_lm(model)
```

```python
from bioinfokit.analys import stat
```

```python
res=stat()
res.tukey_hsd(df_new,res_var="Calories",xfac_var="Milk",anova_model=formula)
```

```python
res.tukey_summary
```

```python
formula="Calories~TotalFat+Sugars+Protein"
lm=ols(formula,df).fit()
lm.summary()
```

```python
df["newcal"]=lm.predict()
```

```python
df["resid"]=df.Calories-df.newcal
```

```python
sns.kdeplot(df.resid) #normal distribution
```
